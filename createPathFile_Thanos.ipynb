```python
## YOLO에 사용할 train.txt 생성 (이미지 경로)
## names/new_txt/max_image 변수 확인
```

```python
import os
import time
import random

darknet_path= '/home/tutor/bpm/darknet/data/'
image_path = '/home/tutor/bpm/dataset/training/'

names = darknet_path + 'obj_final.names' ### 사용하는 .names 파일명
new_txt = darknet_path + 'obj/'+ 'train_final.txt' ### 저장하려는 파일명
# new_txt = 'train_temp.txt' ### 저장하려는 파일명
max_image = int(1000) ## 클래스 당 최대 이미지 개수

with open(new_txt, 'w') as f: 
    with open(names , 'r') as fl: 
        for folder in fl.readlines():
            try:
                folder = folder.strip()
                path = image_path + folder + '/' ## 이미지 파일 있는 경로
                file_list = os.listdir(path)
                random.shuffle(file_list) ## file_list random
                
                count = 0
                start = time.time()
                for file in file_list:
                    try:
                        if (file[-4:] != '.txt') & (file.split('.')[1] != 'ipynb_checkpoints'):
                            count += 1
                            f.write(path + file + '\n') ## 절대경로
                        if count == max_image: break ## 이미지 개수 제한
                    except:
                        print(file, 'file 오류')
                print(folder, '폴더 완료, 실행파일 개수 :', count ,'실행시간 : ', time.time()-start)
            except:
                print(folder, 'open 오류')
```
